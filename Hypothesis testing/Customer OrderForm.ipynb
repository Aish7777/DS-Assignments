```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Costomer+OrderForm.csv")
df.head()
```

```python
df.columns
```

```python
features = ['Phillippines', 'Indonesia', 'Malta', 'India']
```

```python
plt.figure(figsize=(10,8), dpi = 200)

for i in enumerate(features):
    plt.subplot(2,2,i[0]+1)
    sns.countplot(i[1],data = df)
```

```python
chi2, p, dof, expected = stats.chi2_contingency([df['Phillippines'].value_counts(), df['Indonesia'].value_counts(), df['Malta'].value_counts(), df['India'].value_counts()])
print(f'The chi2 value is: {chi2} \nThe p value is: {p}')
if p > 0.05:
    print('Accept null hypothesis')
else:
    print('Reject Null hypothesis')
```

- The p value is greater than 0.05

**Thus, NULL HYPOTHESIS IS ACCEPTED.**
