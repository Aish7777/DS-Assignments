```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Cutlets.csv')
df.head()
```

```python
unitA =df.iloc[:,0]
unitB = df.iloc[:,1]
```

```python
plt.figure(figsize=(10,8), dpi = 90)

plt.hist(unitA,facecolor="c",edgecolor="blue",bins=10)
```

```python
plt.figure(figsize=(10,8), dpi = 90)
plt.hist(unitB, facecolor = 'r', edgecolor = 'y',bins = 10)
```

```python
plt.figure(figsize=(10,8), dpi = 100)
sns.distplot(df['Unit A'])
sns.distplot(df['Unit B'])
plt.legend(['Unit A','Unit B'])
```

```python
stat, p = stats.ttest_ind(unitA,unitB)
```

```python
print(f'The statistic is: {stat} \nThe p value is: {p}')
```

```python
alpha = 0.05
if p > alpha:
    print('Null hypothesis accepted')
else:
    print('Null hypothesis rejected')
```

- The p value is greater than 0.05

**Thus, NULL HYPOTHESIS IS ACCEPTED.**
**There is no significant difference between the diameter of cutlets in 2 units.**
